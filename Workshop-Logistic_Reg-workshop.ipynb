## Logistic Regression Workshop

Load Dataset and Data Cleaning

```python
import pandas as pd
import numpy as np

df = pd.read_csv('iris-data-clean.csv')
display(df)
```

```python
df['class'] = df['class'].replace(['Setosa','Virginica','Versicolor'],[0,1,2])
# left side has to be df['class'] to maintain the dataframe
# otherwise output will be a series

display(df)
```

Model Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

logReg = LogisticRegression()
x= df[['sepal_length_cm', 'sepal_width_cm', 'petal_length_cm', 'petal_width_cm']]
y= df['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
logReg.fit(x_train, y_train)
```

```python
# Provide values to predict the outcome
print(logReg.predict([[6.7, 3.0, 5.2, 2.3]])) #output should be 1
print(logReg.predict([[5.0, 3.6, 1.4, 0.2]])) #output should be 0
```

Validate the Model

```python
# print the confusion matrix and the accuracy score

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

# predict using test data
y_pred = logReg.predict(x_test)

print(accuracy_score(y_test, y_pred))
# first param is always the actual dataset and second is the predicted one

# evaluate using confusion_matrix
print(confusion_matrix(y_test, y_pred, labels=[1,0]))
```

Redo with 2 features!

```python
df_2f = df[['sepal_length_cm','petal_length_cm','class']]
display(df_2f)
```
